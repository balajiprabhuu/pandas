```python
# Merge and Join (Requirement: New data frame, New data, Overlapping columns)

# Concatenate and Append (Requirement: New Rows/Columns)
```

```python
new_df = pd.DataFrame()

new_df["Employee"] = df["Sales Person"].unique()

import random

r = [1,2,4]

random.shuffle(r)

new_df["Experience"] = r

new_df["Total Experience"] = new_df["Experience"] + random.randint(1,10)
```

```python
new_df
```

```python
# To merge we need to mention these parameters (self, left, right, how, on, left_on, right_on, axis, left_index, right_index, sort, suffixes, copy, indicator, validate)

df.merge(new_df, left_on="Sales Person", right_on="Employee")
```

```python
# Since we have two columns for employee name, we can drop the column Employee

# Since we are dropping the column, we need to mention axis=1

df.drop("Employee",axis=1)
```

```python
# Join works based on index not based on column and row names

df.join(new_df)
```

```python
# To overcome this we need set index as sales person on both the data frames

df.reset_index()
df.set_index("Sales Person",inplace=True)
new_df.set_index("Employee",inplace=True)
```

```python
df.join(new_df)
```

```python
#concat joins one df after the other

df1 = df.copy()
df2 = df.copy()
```
